# **Task 1: Zero-Shot Prompting – Leap Year Check**

```python
def is_leap_year(year):
    if (year % 400 == 0) or (year % 4 == 0 and year % 100 != 0):
        return True
    return False
```

```python
print(is_leap_year(2024))
print(is_leap_year(1900))
```

# **Task 2: One-Shot Prompting – Centimeters to Inches**

```python
def cm_to_inches(cm):
    return round(cm / 2.54, 2)
```

```python
print(cm_to_inches(10))
print(cm_to_inches(25))
```

# **Task 3: Few-Shot Prompting – Name Formatting**

```python
def format_name(full_name):
    parts = full_name.strip().split()
    first = parts[0]
    last = parts[-1]
    return f"{last}, {first}"
```

```python
print(format_name("John Smith"))
print(format_name("Anita Rao"))
```

# **Task 4: Comparative Analysis – Zero-Shot vs Few-Shot**

```python
def count_vowels_zero(s):
    count = 0
    for ch in s.lower():
        if ch in "aeiou":
            count += 1
    return count
```

```python
def count_vowels_few(s):
    return sum(1 for ch in s.lower() if ch in "aeiou")
```

```python
print(count_vowels_zero("Artificial Intelligence"))
print(count_vowels_few("Artificial Intelligence"))
```

# **Task 5: Few-Shot Prompting – File Handling**

```python
def count_lines(file_path):
    with open(file_path, 'r') as file:
        return sum(1 for _ in file)
```

```python
with open("sample.txt", "w") as f:
    f.write("Line 1\n")
    f.write("Line 2\n")
    f.write("Line 3\n")
```

```python
print(count_lines("sample.txt"))
```
