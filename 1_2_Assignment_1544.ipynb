# **Task 1: AI-Generated Logic Without Modularization**

```python
n = int(input("Enter a number: "))

result = 1

if n < 0:
    print("Factorial is not defined for negative numbers")
else:
    for i in range(1, n + 1):
        result = result * i

    print("Factorial of", n, "is:", result)
```

# **Task 3: Modular Design Using AI Assistance**

```python
# Task 3: Factorial using modular design and functions

def calculate_factorial(number):
    # This function calculates factorial using iteration
    if number < 0:
        return None

    result = 1
    for i in range(1, number + 1):
        result *= i

    return result


# Main execution block
n = int(input("Enter a number: "))

fact = calculate_factorial(n)

if fact is None:
    print("Factorial is not defined for negative numbers")
else:
    print(f"Factorial of {n} is: {fact}")
```

# **Task 5: AI-Generated Iterative vs Recursive Thinking**

```python
# Iterative factorial using loop

def factorial_iterative(n):
    if n < 0:
        return None

    result = 1
    for i in range(1, n + 1):
        result *= i

    return result


num = int(input("Enter a number: "))
print("Iterative Factorial:", factorial_iterative(num))
```

```python
# Recursive factorial implementation

def factorial_recursive(n):
    if n < 0:
        return None
    if n == 0 or n == 1:
        return 1
    return n * factorial_recursive(n - 1)


num = int(input("Enter a number: "))
print("Recursive Factorial:", factorial_recursive(num))
```
