# **Task-1:Zero-Shot Prompting – Palindrome Number Program**

```python
def is_palindrome(number):
    """
    This function checks whether a given number is a palindrome.
    Returns True if palindrome, otherwise False.
    """
    if not isinstance(number, int):
        return False

    if number < 0:
        return False

    original = number
    reverse = 0

    while number > 0:
        digit = number % 10
        reverse = reverse * 10 + digit
        number //= 10

    return original == reverse


test_numbers = [121, 123, 0, 9, 10, -121, 1221]

for num in test_numbers:
    print(f"Input: {num} -> Palindrome: {is_palindrome(num)}")
```

# **Task-2:One-Shot Prompting – Factorial Calculation**

```python
def factorial(n):
    """
    This function calculates the factorial of a given number.
    Returns factorial value or error message.
    """
    if not isinstance(n, int):
        return "Invalid input: Not an integer"

    if n < 0:
        return "Factorial not defined for negative numbers"

    result = 1
    for i in range(1, n + 1):
        result *= i

    return result

test_values = [5, 0, 1, 7, -3, 4.5]

for value in test_values:
    print(f"Input: {value} -> Output: {factorial(value)}")
```

# **Task-3:Few-Shot Prompting – Armstrong Number Check**

```python
def is_armstrong(number):
    """
    Checks whether the given number is an Armstrong number.
    """
    if not isinstance(number, int):
        return "Invalid Input"

    if number < 0:
        return "Invalid Input"

    digits = list(map(int, str(number)))
    power = len(digits)

    total = 0
    for digit in digits:
        total += digit ** power

    if total == number:
        return "Armstrong Number"
    else:
        return "Not an Armstrong Number"


test_numbers = [153, 370, 371, 123, 0, 1, -153, "abc"]

for num in test_numbers:
    print(f"Input: {num} -> Output: {is_armstrong(num)}")
```

# **Task-4:Context-Managed Prompting**

```python
def classify_number(n):
    """
    Classifies a number as Prime, Composite, or Neither.
    """
    if not isinstance(n, int):
        return "Invalid Input"

    if n <= 1:
        return "Neither Prime nor Composite"

    for i in range(2, int(n ** 0.5) + 1):
        if n % i == 0:
            return "Composite"

    return "Prime"


test_values = [2, 3, 4, 9, 1, 0, -5, 17, 18, 2.5]

for value in test_values:
    print(f"Input: {value} -> Classification: {classify_number(value)}")
```

# **Task-5:Zero-Shot Prompting – Perfect Number Check**

```python
def is_perfect_number(num):
    """
    Checks whether a given number is a perfect number.
    """
    if not isinstance(num, int):
        return "Invalid Input"

    if num <= 0:
        return False

    divisor_sum = 0

    for i in range(1, num):
        if num % i == 0:
            divisor_sum += i

    return divisor_sum == num


test_numbers = [6, 28, 12, 1, 496, -6, "hello"]

for num in test_numbers:
    print(f"Input: {num} -> Perfect Number: {is_perfect_number(num)}")
```

**Task-6:Few-Shot Prompting – Even or Odd with Validation**

```python
def even_or_odd(number):
    """
    Determines whether a number is Even or Odd with validation.
    """
    if not isinstance(number, int):
        return "Invalid Input"

    if number % 2 == 0:
        return "Even"
    else:
        return "Odd"


test_values = [8, 15, 0, -4, -7, 3.14, "abc"]

for value in test_values:
    print(f"Input: {value} -> Output: {even_or_odd(value)}")
```
